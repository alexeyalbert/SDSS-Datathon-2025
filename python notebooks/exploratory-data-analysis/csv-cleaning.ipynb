```python
import pandas as pd
df = pd.read_csv('./Transportation-Data/subway-data.csv')

df_cleaned = df[~df['Station'].str.contains(' to ', case=False)].copy()

# checking numnber of unique stations
unique_stations = sorted(df_cleaned['Station'].unique())
for station in unique_stations:
    print(station)
print(len(unique_stations))

unique_stations = pd.DataFrame(unique_stations, columns=['Column_Name'])


unique_stations.to_csv('./Transportation-Data/unique_stations.csv', index=False)
```

```python
# put the data into a nice format for mapping
df_replacements = pd.read_csv('/Users/xiaotongshen/Desktop/SDSS-Datathon-2025/python notebooks/exploratory-data-analysis/Transportation-Data/subway-stations.csv', header=None)

# Create a replacement mapping dictionary
replacement_mapping = {}

# Iterate through the DataFrame to populate the mapping
for index, row in df_replacements.iterrows():
    # Split the string by the last occurrence of ' - '
    parts = row[0].rsplit(' - ', 1)  # Split from the right, only once
    if len(parts) == 2:  # Ensure there are exactly two parts
        original, new = parts  # Unpack the parts
        replacement_mapping[original.strip()] = new.strip()  # Add to the dictionary
# Display the replacement mapping
print(replacement_mapping)

df_cleaned['Station'] = df_cleaned['Station'].replace(replacement_mapping)
print(df_cleaned)

# remove n/a or na values
df_cleaned = df_cleaned[~df_cleaned['Station'].isin(['N/A', 'NA'])]
print(df_cleaned)
```
