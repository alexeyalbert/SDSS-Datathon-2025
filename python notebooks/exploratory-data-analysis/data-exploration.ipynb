```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./Transportation-Data/subway-data.csv')

df_cleaned = df[~df['Station'].str.contains(' to ', case=False)].copy()
```

```python
# Group data by station and calculate delay probability
station_delays = df.groupby('Station')['Min Delay'].mean().sort_values(ascending=False)

print(station_delays)
# Create a bar plot of the top 15 stations with highest delay probability
plt.figure(figsize=(12, 6))
sns.barplot(x=station_delays.head(15).index, y=station_delays.head(15).values)
plt.xticks(rotation=45, ha='right')
plt.title('Top 15 Stations with Highest Average Delay ')
plt.xlabel('Station')
plt.ylabel('Delay Probability')
plt.tight_layout()
plt.show()

# Display the numerical results
print("\nTop 15 stations with highest delay probability:")
print(station_delays.head(15))
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better visualization
# plt.style.use(sns)

# Read the data
df = pd.read_csv('./Transportation-Data/subway-data.csv')


# Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Min Delay', bins=30, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Distribution of Min Delays Across Stations', fontsize=14, pad=15)
plt.xlabel('Min Delays (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()

# Print summary statistics
print("\nSummary Statistics of Min Delays:")
print(df['Min Delays'].describe())
```
